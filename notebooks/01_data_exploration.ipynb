# 01 — Data Exploration
Load sample water data and visualize parameter distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/samples/sample_input.csv')
print(df.shape)
df.head()
```

```python
df.describe().T
```

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 10))
params = ['pH','TDS','hardness','turbidity','nitrates','fluoride','chlorides','dissolved_oxygen','BOD','coliform']
for i, col in enumerate(params):
    df[col].hist(ax=axes[i//4][i%4], bins=10, color='steelblue', edgecolor='white')
    axes[i//4][i%4].set_title(col)
plt.tight_layout()
plt.show()
```
