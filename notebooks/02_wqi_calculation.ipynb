# 02 — WQI Calculation
Run the WQI engine on sample data using BIS and WHO standards.

```python
import sys
sys.path.insert(0, '../src')
import pandas as pd
from engine.wqi_calculator import calculate_wqi

df = pd.read_csv('../data/samples/sample_input.csv')
```

```python
param_cols = ['pH','TDS','hardness','turbidity','nitrates','fluoride','chlorides','dissolved_oxygen','BOD','coliform']
results = []
for _, row in df.iterrows():
    sample = row[param_cols].to_dict()
    res = calculate_wqi(sample, standard_name='bis')
    res['sample_id'] = row['sample_id']
    res['location'] = row['location']
    results.append(res)

result_df = pd.DataFrame([{'sample_id':r['sample_id'],'location':r['location'],'wqi':r['wqi'],'classification':r['classification']} for r in results])
result_df
```

```python
import matplotlib.pyplot as plt
result_df.plot.bar(x='sample_id', y='wqi', color='coral', edgecolor='black')
plt.axhline(50, color='orange', linestyle='--', label='Good threshold')
plt.axhline(100, color='red', linestyle='--', label='Unsuitable threshold')
plt.legend()
plt.title('WQI by Sample (BIS Standard)')
plt.tight_layout()
plt.show()
```
