# 03 — Treatment Recommendations
Generate treatment advice for each water sample.

```python
import sys
sys.path.insert(0, '../src')
import pandas as pd
from treatment.recommender import get_recommendations

df = pd.read_csv('../data/samples/sample_input.csv')
param_cols = ['pH','TDS','hardness','turbidity','nitrates','fluoride','chlorides','dissolved_oxygen','BOD','coliform']
```

```python
for _, row in df.iterrows():
    sample = row[param_cols].to_dict()
    recs = get_recommendations(sample)
    print(f"\n--- {row['sample_id']} ({row['location']}) ---")
    for r in recs:
        print(f"  [{r['priority']}] {r['parameter']}: {r['recommendation']}")
```
